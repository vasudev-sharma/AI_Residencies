```
import pandas as pd
import numpy as np
```

```
df = pd.DataFrame(np.random.random((3, 3)), index=list('abc'))
```

```
df.to_csv('test.csv') # save the pandas dataframe. By default, it saves with index
```

```
df
```

```
pd.read_csv('test.csv') 
# Note while loading the csv file, the index is also added.
```

```
# To prevent the above behaviour, use `index` = False flag
df.to_csv('test.csv', index=False)
# NOTE: always use index arg while saving a csv file
```

```
pd.read_csv('test.csv')
```
